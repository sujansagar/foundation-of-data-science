```python

differences = np.zeros(5000)
```

```python
for i in np.arange(5000):
  newdf = df[['GENDER', 'PRICE']]
  shuffled = newdf.sample(12491,replace = False)
  shuffled_volume = shuffled['PRICE']
  original_and_shuffled = newdf.assign(shuffled_volume=shuffled_volume.values )
  all_group_means= original_and_shuffled.groupby('GENDER').mean()
  difference = all_group_means['shuffled_volume'][0]- all_group_means['shuffled_volume'][1]
  differences[i]=difference
```

```python
differences_df = pd.DataFrame(differences)
differences_df
```

```python
differences_df.hist(bins = 15)
plt.title('Prediction that is overNull Hypotheses');
plt.xlabel('changes over Group Averages',fontsize=14)
plt.ylabel('Units',fontsize=23);
plt.scatter(observed_difference,123,color='blue',s=20)
```

```python
P_Value = np.count_nonzero(differences <= observed_difference)/differences.size
P_Value
```

since p value is small ,so we reject null hypothesis
