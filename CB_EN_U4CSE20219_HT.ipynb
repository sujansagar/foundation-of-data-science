# Hypothesis testing

## NULL HYPOTHESIS: gender women makes average PRICE of 1356
## ALTERNATIVE HYPOTHESIS: gender women average PRICE more than of 1356 

```python
sample=df.loc[df['GENDER']=='Women']
newsamplele=sample.sample(187)
newsample
```

```python
newsample.PRICE.mean()
```

## z test

```python

sd=np.std(newsample['PRICE'])
sd
```

```python
newsample.std()
```

```python
xbar=newsample.PRICE.mean()
mu=1356
n=189 
alpha=0.05
```

```python
xbar
```

```python
zvalue=(xbar-mu)/(sd/np.sqrt(n))
zvalue
```

```python
z_critical = st.norm.ppf(alpha)  
z_critical
```

```python
if (zvalue<z_critical and abs(zvalue)> abs(z_critical)):
  print("Reject Null Hypotheisis")
else:
  print("cant reject Null Hypothesis")
```

## using p value

```python
p_val = (1 - st.norm.cdf(abs(zvalue)))
p_val
```

```python

if p_val<alpha:
  print("Reject Null Hypotheisis")
else:
  print("Fail to reject Null Hypothesis")
```

## Hence GENDER Women On mean doesn't make more than 1356 per day so thatwe fail to reject the Null Hypothesis

## T test

```python
TSAMPLE=sample.sample(34)
TSAMPLE
```

```python
t_sd=np.std(TSAMPLE['PRICE'])
t_sd
```

```python
xbar=TSAMPLE.PRICE.mean()
print(xbar)
mu=1356
n=34
alpha=0.05
```

```python
n = 34
degrees_of_freedom = n-1
t = (xbar - mu)/(t_sd / np.sqrt(n))
t
```

```python
t_critical = st.t.ppf(alpha, degrees_of_freedom)
t_critical
```

```python
if (t > t_critical):
    print("Null hypothesis cannot be rejected")
else:
    print("Reject null hypothesis")
```

## so that both Z test and T test gives same result, we cant reject null hypothesis.
